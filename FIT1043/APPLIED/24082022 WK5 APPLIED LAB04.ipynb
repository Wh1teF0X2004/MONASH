# Introduction to the Concept of Linear Regression

```python
random.seed(123)
indVar=random.sample(range(1, 30), 20)
indVar
# define a linear function to represent the general pattern between dependent and independent variables
```

```python
# depVar = indVar
depVar = indVar[:]
depVar
```

```python
plt.scatter(indVar,depVar)
plt.show()
```

```python
plt.scatter(indVar,depVar)
plt.plot(indVar,depVar,'r-', linewidth=3) # red line, of thickness 3
plt.show()
```

```python
trueX = indVar[:]
trueY = depVar[:]
```

```python
noiseXVals=random.sample(range(1,30),10)
indVar.extend(noiseXVals)
```

```python
noiseYVals=[x+round(random.uniform(-10,6)) for x in noiseXVals]
depVar.extend(noiseYVals)
```

```python
plt.scatter(indVar,depVar)
# the red line represents the true data
plt.plot(trueX,trueY,'r-', linewidth=3)
plt.show()
```

```python
noiseXVals=random.sample(range(1,30),25)
indVar.extend(noiseXVals)
noiseYVals=[x+round(random.uniform(-10,6))for x in noiseXVals]
depVar.extend(noiseYVals)
plt.scatter(indVar,depVar)
plt.plot(trueX,trueY,'r-', linewidth=3)
plt.show()
```

```python
trueYNew=[x-1.5 for x in trueY]
plt.scatter(indVar,depVar)
plt.plot(trueX,trueY,'r-', linewidth=3)
plt.plot(trueX,trueYNew,'g-', linewidth=3)
plt.show()
```

# Linear Regression in Python

```python
# Scientific computation purposes
from scipy.stats import linregress
# have modules to import linear regressions
```

```python
slope, intercept, r_value, p_value, std_err = linregress(indVar,depVar)
```

```python
print("slope: %f intercept: %f" % (slope, intercept))
print("r-value: %f" % r_value) 
# r = core relation between both axis (higher = more relation)
print("p-value: %f" % p_value) # p = used for hypothesis testing
```

```python
line = [slope*xi + intercept for xi in indVar]
# our 'raw' data
plt.scatter(indVar, depVar)
# our earlier manual estimated line (in green)
plt.plot(trueX,trueYNew,'g-', linewidth=1)
# the line created by the linregress() function (in red)
plt.plot(indVar,line,'r-', linewidth=1)
plt.show()
```

```python
line = [slope*xi + intercept for xi in indVar]
# our 'raw' data
plt.scatter(indVar, depVar)
# our earlier manual estimated line (in dot green)
plt.plot(trueX,trueYNew,'dg-', linewidth=1)
# the line created by the linregress() function (in dot red)
plt.plot(indVar,line,'dr-', linewidth=1)
plt.show()
```

```python
# Practice 1:
df = pd.DataFrame(
{'Name' : ['Mike','Aaron','Brad','Steve','George','Mitchell',

'Shaun','Glenn','Pat','Robert','David'],
'Age' : [22, 25, 25, 26, 28, 28, 31, 32, 33, 39, 44],
'Runs' :[1000, 980, 900, 900, 828, 700, 672, 662, 655, 600, 557]}
)
df
```

```python
sns.scatterplot(data=df, x="Age", y="Runs", palette ="pastel").set(title="Age vs Run in between 11 people")
```

```python
slope, intercept, r_value, p_value, std_err = linregress(df["Age"],df["Runs"])
```

```python
line = [slope*xi + intercept for xi in df["Age"]]

plt.plot(df["Age"],line,'r-', linewidth=2)
plt.scatter(df["Age"], df["Runs"])

plt.show()
```

```python
print("slope: %f intercept: %f" % (slope, intercept))
print("r-value: %f" % r_value) 
print("p-value: %f" % p_value)
```
